```python
from OCR.IO import read_languages
from OCR.image_gen.RNG import RandomTextGenerator, RandomTextImageGenerator
size = (int(1387), int(445))
new_size = (int(1387/5), int(445/5  ))
max_chars = 8

languages = read_languages()
text_rng = RandomTextGenerator(
    languages, min_chars=7, max_chars=max_chars, max_actual_height=0,
    fixed_size=True, max_actual_width=size[0] / max_chars - 5
)
img_rng_gen = RandomTextImageGenerator(size, text_rng)()
img = next(img_rng_gen).img
img
```
